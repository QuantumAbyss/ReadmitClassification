```python
datClean = dat.drop(['medical_specialty', 'diag_1', 'diag_2', 'diag_3', 'encounter_id', 'patient_nbr'], axis=1)
datClean.info()
```

```python
datClean = pd.get_dummies(datClean,drop_first=True)
datClean.head()
```

```python
from sklearn.model_selection import train_test_split

X = datClean.drop('readmitted_NO',axis=1)
Y = datClean['readmitted_NO']
Xtrain, Xvalid, Ytrain, Yvalid = train_test_split(X, Y, test_size=0.30)
```

```python
Xtrain.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfmodel = RandomForestClassifier(n_estimators=600)
rfmodel.fit(Xtrain,Ytrain)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, roc_auc_score
predictions = rfmodel.predict(Xvalid)
print(classification_report(Yvalid,predictions))
```

```python
print(confusion_matrix(Yvalid, predictions))
```

```python
roc_auc_score(Yvalid, rfmodel.predict_proba(Xvalid)[:,1])
```

```python
importance = rfmodel.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
CrosstabResult=pd.crosstab(index=dat['medical_specialty'],columns=dat['readmitted'])
print(CrosstabResult)
```

Higher percentage indicates higher likelihood of readmittance.

```python
toPlot = CrosstabResult
toPlot['medical_specialty'] = toPlot.index
toPlot['percent'] = 100*(toPlot['<30'] / (toPlot['<30'] + toPlot['NO']))
toPlot.head()

dims = (16, 8)
fig, ax = pyplot.subplots(figsize=dims)
barPlot = sns.barplot(x='medical_specialty',y='percent', ax=ax, data=toPlot)
pyplot.setp(barPlot.get_xticklabels(), rotation=90)
barPlot
```

Chisquare below out of curiosity

```python
# importing the required function
from scipy.stats import chi2_contingency

#Error with '<' symbol in the column name
CrosstabResult=pd.crosstab(index=dat['medical_specialty'],columns=dat['readmitted'])
CrosstabResult.columns = ['lessthan30', 'NO']

# Performing Chi-sq test
ChiSqResult = chi2_contingency(CrosstabResult)
 
# P-Value is the Probability of H0 being True
# If P-Value>0.05 then only we Accept the assumption(H0)
 
print('The P-Value of the ChiSq Test is:', ChiSqResult[1])
```
